# 1 自编逐个计算距离KNN

```python
def Distance(x,y,p):
    '''
    返回x到y的p阶矩
    '''
    return np.power(np.sum(np.power(np.abs(x-y),p)),1/p)
```

```python
x1 = np.array([1, 1],dtype=np.float64)
x2 = np.array([5, 1],dtype=np.float64)
x3 = np.array([4, 4],dtype=np.float64)
```

```python
for i in range(1,5):
    r = {'{}-{}'.format(i,c):Distance(x1,c,p=i) for c in [x2,x3]}
    print(r)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['label'] = iris.target
df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'label']
```

```python
df.head()
```

```python

plt.scatter(df[:50]['sepal length'], df[:50]['sepal width'], label='0')
plt.scatter(df[50:100]['sepal length'], df[50:100]['sepal width'], label='1')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()
```

```python
data = np.array(df.iloc[:100, [0, 1, -1]])
X, y = data[:,:-1], data[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score
class KNN:
    def __init__(self,X_train,Y_train,n_neighbors = 3,p = 2):
        '''
        n_neighbors:临近点个数
        p:距离计算方式
        '''
        self.n_neibors = n_neighbors
        self.p = p
        self.X_train = X_train
        self.Y_train = Y_train
    
    def predict(self,X):
        '''
        预测X的类别
        '''
        min_dis = []#存储最近的n个点的距离和类标
        
        dist = np.linalg.norm(self.X_train - X,ord = self.p,axis = 1)

        min_dis_index = np.argsort(dist)[:self.n_neibors]

        y_pred = Counter(self.Y_train[min_dis_index]).most_common(1)[0][0]

        return y_pred

    def score(self,X_test,Y_test):
        '''
        计算预测准确率
        '''
        y_pred = []
        for i in X_test:
            y_pred.append(self.predict(i))
        return accuracy_score(Y_test,np.array(y_pred))
        
        
```

```python
clf = KNN(X_train, y_train)
clf.score(X_test,y_test)
```

```python
test_point = [6.0, 3.0]
print('Test Point: {}'.format(clf.predict(test_point)))
```
